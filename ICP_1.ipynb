```python
# import libraries
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/hepatitis/master/data/hepatitis.csv')
df.head(5)
```

```python
df.shape
```

###**Replace the missing values in numeric columns of  dataset (hepatitis dataset) with the mean values of the column.**

```python
# Select the numeric columns at first.
numeric_cols = df.select_dtypes(include=np.number).columns
numeric_cols
```

```python
# print the data types of the numeric columns
for col in numeric_cols:
  print(col, df[col].dtypes, sep='   ')
```

```python
# find the number of missing values in the numeric columns
for col in numeric_cols:
  print(col, df[col].isna().sum(), sep='   ')
```

```python
# mean of each numeric column in the dataset
for col in numeric_cols:
  print(col, df[col].mean(), sep='   ')
```

### **Replace missing values of numeric columns with the mean of those columns**

```python
for col in numeric_cols:
  df[col].fillna(df[col].mean(), inplace=True)
df.head(5)
```

```python
# Check if there are still any missing values
for col in numeric_cols:
  print(col, df[col].isna().sum(), sep='   ')
```
